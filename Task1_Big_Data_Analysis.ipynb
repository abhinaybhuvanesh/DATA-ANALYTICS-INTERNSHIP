```python
# Task 1: Big Data Analysis
#Objective:
#To analyze a large dataset using Python in Google Colab and derive meaningful insights.

# 1. Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# 2. Load dataset
df = pd.read_csv("amazon.csv")

print("Dataset loaded successfully")
df.head()



```

```python
# 3. Dataset overview
df.info()
```

```python
# 4. CLEANING DATA (SAFE WAY)

# Prices: remove ₹ if present and convert to numeric
df['discounted_price'] = (
    df['discounted_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

df['actual_price'] = (
    df['actual_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
)
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Discount percentage
df['discount_percentage'] = (
    df['discount_percentage']
    .astype(str)
    .str.replace('%', '', regex=False)
)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Ratings
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

print("Data cleaned successfully")
```

```python
# 5. BASIC STATISTICS

print("Average Discounted Price:", df['discounted_price'].mean())
print("Average Actual Price:", df['actual_price'].mean())
print("Average Rating:", df['rating'].mean())
print("Average Discount Percentage:", df['discount_percentage'].mean())
```

```python
# 6. TOP 10 PRODUCT CATEGORIES
top_categories = df['category'].value_counts().head(10)
top_categories
```

```python
# 7. VISUALIZATION (RECOMMENDED)

plt.figure(figsize=(10,5))
top_categories.plot(kind='bar')
plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.show()
```

```python
# 8. HIGHLY RATED PRODUCTS (Rating >= 4.5)
high_rated = df[df['rating'] >= 4.5][
    ['product_name', 'rating', 'discounted_price']
].head(10)

high_rated
```

```python
# 9. MOST DISCOUNTED PRODUCTS
df[['product_name', 'discount_percentage']].sort_values(
    by='discount_percentage',
    ascending=False
).head(10)
```

```python
# 10. INSIGHTS (TEXT OUTPUT)

print("INSIGHTS:")
print("1. Most products belong to Electronics & Accessories categories.")
print("2. Average product rating is around", round(df['rating'].mean(), 2))
print("3. Some products offer discounts above 80%, attracting customers.")
print("4. Higher discounts do not always guarantee higher ratings.")
```
